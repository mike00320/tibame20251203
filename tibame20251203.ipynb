````python
# 判斷此年是否為閏年
# 2000 閏年
# 1800 平年
# 1804 閏年
# 1803 平年

#第一種 (我)
year = 1803
if year % 4 != 0:
    print("平年")
elif year % 100 != 0:
    print("閏年")
elif year % 400 != 0:
    print("平年")
else:
    print("閏年")

#第二種
year = 1803

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("O")
        else:
            print("X")
    else:
        print("O")
else:
    print("X")

#第三種
year = 2000
if year % 400 == 0:
    print("O")
elif year % 100 == 0:
    print("X")
elif year % 4 == 0:
    print("O")
else:
    print("X")
````

````python
import ColabTurtle.Turtle as turtle

turtle.initializeTurtle()

turtle.speed(6)
init_width = 40
shape_count = 6
shape_angle = 360 / shape_count

prev_angle = 0

i = 0
while i < shape_count:
    turtle.forward(init_width)
    x, y = turtle.pos()
    print(x, y)
    turtle.left(180)
    turtle.forward(init_width)
    turtle.left(180 - shape_angle)
    turtle.forward(init_width)
    turtle.left(180 - ((180-shape_angle) / 2))
    turtle.goto(x,y)
    turtle.left(180 - ((180-shape_angle) / 2))
    turtle.forward(init_width)
    turtle.left(180-shape_angle)
    i = i + 1

````

````python
import ColabTurtle.Turtle as turtle

turtle.initializeTurtle()

turtle.speed(8)
init_width = 40
shape_count = 10
shape_angle = 360 / shape_count

prev_angle = 0

i = 0
while i < shape_count:
    startx, starty = turtle.pos()
    turtle.forward(init_width)
    prevx, prevy = turtle.pos()
    print(prevx, prevy)
    turtle.goto(startx, starty)
    turtle.right(shape_angle)
    turtle.forward(init_width)
    turtle.goto(prevx, prevy)
    turtle.goto(startx, starty)
    i = i + 1
````
